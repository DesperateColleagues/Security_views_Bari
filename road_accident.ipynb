## Geocoding
Find the district for each road accident

### Merging district's Polygons

```python
import geopandas as gpd
import matplotlib.pyplot as plt

from geopy import Nominatim
from geopy.exc import GeocoderTimedOut
```

```python
quartieri_sf = gpd.read_file("Raw Datasets/Quartieri_Bari/Quartieri.shp")
```

```python
quartieri_sf = (
    quartieri_sf
        .groupby(by='nome')
        .apply(func=lambda x: x.dissolve())['geometry'] # used to work on sub DF
        .reset_index()
        .drop(columns='level_1')
)
```

```python
quartieri_sf = quartieri_sf.to_crs(4326)
```

```python
quartieri_sf
```

### Transform accident lat&log in GeoPoint

```python
geodf_cars = gpd.GeoDataFrame(
    df_cars,
    geometry=gpd.points_from_xy(df_cars['Longitudine'], df_cars['Latitudine']), crs="EPSG:4326"
)
```

### Name of the accident district

```python
# List of districts that contain the points
lst = []
# District geometries
qs = quartieri_sf['geometry']

for i, p in enumerate(geodf_cars['geometry']):

    quartieri_contains_pt = quartieri_sf[qs.contains(p)]['nome'].to_list()
    # If there is at least a district that contains the point p
    if len(quartieri_contains_pt) != 0:
        lst.append(quartieri_contains_pt[0])
    else:
        lst.append('NOT FOUND')
```

```python
lst
```

#### Drop accidents that occurred outside Bari

```python
accidents_not_found_list = []

# Individuo l'indice dei punti fuori da bari
for i, s in enumerate(lst):
    if (s == 'NOT FOUND'):
        accidents_not_found_list.append(i)
```

```python
mask_index = [ i for i in geodf_cars.index if i not in accidents_not_found_list]
```

```python
# Select only the accidents in Bari
geodf_cars = geodf_cars.loc[ mask_index, :]
```

### Map district's geometry

```python

district_names_list = np.asarray(lst)[mask_index]
geodf_cars['District name'] = district_names_list
geodf_cars['District geometry'] = list(map( lambda name: quartieri_sf[quartieri_sf['nome']==name].geometry.values[0], district_names_list))
```

### Line chart

Import original datasets

```python
accidents_2022 = pd.read_csv("Raw Datasets/Elenco_incidenti_2022.csv", sep=';')
accidents_2023 = pd.read_csv("Raw Datasets/Elenco_incidenti_2023.csv", sep=',')
```

Extract the month for each road accident

```python
temp2022 = accidents_2022.copy()
temp2023 = accidents_2023.copy()

temp2022['Mese incidente'] = accidents_2022['Data incidente'].map(lambda s: s.split('/')[1])
temp2023['Mese incidente'] = accidents_2023['Data incidente'].map(lambda s: s.split('/')[1])
```

Count road accidents for month

```python
temp2022 = temp2022.groupby('Mese incidente')['lat'].count().reset_index().rename(columns={'lat': 'n_incidenti_2022'})
temp2023 = temp2023.groupby('Mese incidente')['Latitudine'].count().reset_index().rename(columns={'Latitudine': 'n_incidenti2023'})
```

```python
accidents = temp2022.merge(temp2023, on="Mese incidente")
```

Map the month's name for each road accident

```python
months_dict = {
    "01": "Gennaio",
    "02": "Febbraio",
    "03": "Marzo",
    "04": "Aprile",
    "05": "Maggio",
    "06": "Giugno",
    "07": "Luglio",
    "08": "Agosto",
    "09": "Settembre",
    "10": "Ottobre",
    "11": "Novembre",
    "12": "Dicembre",
}
```

```python

accidents['Mese incidente'] = accidents['Mese incidente'].map(months_dict)
```

Save new Dataset

```python
accidents.to_csv("./Fluorish/modified_Road_accidents_lineChart.csv")
```

### Parallel coordinates Race

```python
geodf_cars.head(1)
```

```python
race_df =  geodf_cars.loc[:, ['Data incidente', 'Caratteristiche della strada']]
```

Map the month's name for each road accident

```python
months_dict = {
    "01": "Gennaio",
    "02": "Febbraio",
    "03": "Marzo",
    "04": "Aprile",
    "05": "Maggio",
    "06": "Giugno",
    "07": "Luglio",
    "08": "Agosto",
    "09": "Settembre",
    "10": "Ottobre",
    "11": "Novembre",
    "12": "Dicembre",
}

race_df['Month'] = race_df['Data incidente'].map(lambda s: s.split('/')[1]).map(months_dict)
race_df.head()
```

Update categories of the type roads

```python
Type_road_dict = {
    "Rettilineo": "Rettilineo",
    "Pianeggiante": "Rettilineo",
    "Cavalcavia": "Rettilineo",
    "Strettoia": "Rettilineo",
    "Galleria illuminata": "Rettilineo",
    "Pendenza": "Rettilineo",
    "Sottopasso illuminato": "Rettilineo",
    "Sottopasso non illuminato": "Rettilineo",

    "Intersezione semaforizzata": "Intersezione semaforizzata",

    "Rotatoria": "Rotatoria",
    
    "Incrocio": "Incrocio",
    "Intersezione stradale segnalata": "Incrocio",
    "Intersezione non regolata/non segnalata": "Incrocio",
    "Passaggio a livello (custodito)": "Incrocio",    

    "Curva a visuale libera": "Curva",
    "Curva senza visuale libera": "Curva"
}

race_df['Caratteristiche della strada'] = race_df['Caratteristiche della strada'].map(Type_road_dict)
race_df.head()
```

```python
race_df = race_df.pivot_table(index='Caratteristiche della strada', columns='Month', values='Month', aggfunc="size").reset_index()
```

Replace NA values with 0 (if present)

```python
# Rows with at least one NA
row_filter = race_df.isna().sum(axis=1)
race_df.loc[row_filter > 0, :]
```

```python
race_df.fillna(value=0)
```

Sort the month by temporal order

```python
columns_order = [
    "Caratteristiche della strada", 
    "Gennaio", "Febbraio", "Marzo", "Aprile", "Maggio", "Giugno",
    "Luglio", "Agosto", "Settembre", "Ottobre", "Novembre", "Dicembre"]

race_df = race_df.loc[:, columns_order]
```
