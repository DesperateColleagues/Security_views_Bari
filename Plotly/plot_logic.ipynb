```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df_cars_jak = pd.read_csv('2023_car_crash_paralle_sets.csv')
```

```python
mapping = {
    'Strada Urbana': 0,
    'Strada extraurbana': 1
}
```

```python
# Create dimensions
dim1 = go.parcats.Dimension(
    values=df_cars_jak['Stato fondo stradale'],
    label="Stato fondo stradale",
    #categoryorder='category descending'
)

dim2 = go.parcats.Dimension(
    values=df_cars_jak['Tipo di strada'], 
    label="Tipo di strada",
)

dim3 = go.parcats.Dimension(
    values=df_cars_jak['Condizioni traffico'],
    label='Condizioni di traffico',
    categoryorder='array',
    categoryarray=['Intenso', 'Normale', 'Scarso']
)

# Create parcats trace
color = df_cars_jak['Tipo di strada'].map(mapping);
colorscale = [[0, '#006E90'], [1, '#F18F01']];

fig = go.Figure(
    data = [go.Parcats(dimensions=[dim2, dim1, dim3],
        line={'color': color, 'colorscale': colorscale},
        hoveron='color', hoverinfo='count',
        labelfont={'size': 16, 'family': 'Arial',},
        tickfont={'size': 13, 'family': 'Arial'},
        arrangement='freeform')],
    )


fig.update_layout(
    autosize=False,
    margin=dict(l=100, r=80, t=80, b=50),
)

fig.update_yaxes(automargin='left+top')


fig.show()
```

```python
fig.write_html('plot.html')
```
